```python
tableName = "dataset_TEST"
tableName = tableName.lower()

####문자컬럼에 대해서 varchar (100) 적용 _ 사용시 속도 SpeedUp 50배 
objectColumns = list(df.columns[df.dtypes == 'object']) #타입을 오브젝트로 varchar (100)
typeDict={}
maxLen = 100

for i in range(0, len(objectColumns)):
    #sawonData[i].str.len().max() 최대치 사용할 경우
        typeDict[ objectColumns[i] ] = types.VARCHAR(100)
####문자컬럼에 대해서 varchar (100) 적용 _ 사용시 속도 SpeedUp 50배 

df.to_sql(name=tableName, if_exists="replace", con=engine,dtype=typeDict,index=False )
```

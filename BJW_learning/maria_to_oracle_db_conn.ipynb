```python
import pandas as pd
import psycopg2
from sqlalchemy import create_engine,types
```

```python
##Maria DB
inId = "kopo"
inPw = "kopo"
inIp = "192.168.110.111"
inPort= "3306"
inDbName = "kopo"
```

```python
#conn
postEngine = create_engine('mysql+pymysql://{}:{}@{}:{}/{}'.format(inId, inPw, inIp , inPort ,inDbName) )
```

```python
conDB = pd.read_sql_query(sql="select * from kopo_bankdata", con = postEngine)
```

```python
conDB
```

```python
groupKey = ["region"]

quizData = conDB.groupby( groupKey )[["income"]].mean().reset_index()

quizData.rename(columns = {"income": "income_mean"}, inplace =True)

quizData
```

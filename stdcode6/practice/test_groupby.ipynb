```python
import pandas as pd
import numpy as np
```

```python
idx=['A','A','B','B','B','C','C','C','D','D','D','D','E','E','E']
col=['col1','col2','col3']
data = np.random.randint(0,9,(15,3))
df = pd.DataFrame(data=data, index=idx, columns=col).reset_index()
```

```python
df
```

```python
df.groupby('index').mean() #mean()평균 계산
```

```python
df.groupby('index').count() #count() 카운트
```

```python
df.groupby('index').agg(['sum','mean']) #agg메서드 여러 연산 가능
```

```python
df.groupby('index',group_keys=False).apply()   #apply( 함수, axis = 0 or 1 )
```
